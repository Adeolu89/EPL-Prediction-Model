```python
columns_to_drop= ['Referee', 'Attendance', 'Match Report','Notes']
epl_df.drop(columns = columns_to_drop, inplace=True)

# Dropping these columns as they are not going to be useful for our analysis
```

```python
epl_df['Date'] = pd.to_datetime(epl_df['Date'])
```

```python
epl_df["venue_code"] = epl_df['Venue'].astype("category").cat.codes
```

```python
epl_df["opp_code"] = epl_df["Opponent"].astype("category").cat.codes
epl_df
```

```python
epl_df["Target"] = (epl_df["Result"] == "W").astype("int")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
train = epl_df[epl_df["Date"] < '2024-01-01']
test = epl_df[epl_df["Date"] >= '2024-01-01']
```

```python
predictors = ["venue_code", "opp_code", "Day_Code"]
```

```python
features_train = train[predictors]
labels_train = train["Target"]
features_test = test[predictors]
labels_test = test["Target"]
```

```python
rf.fit(features_train, labels_train)
```

```python
preds = rf.predict(features_test)
```

```python
acc = accuracy_score(preds, labels_test)
print(acc)
```

```python
combined = pd.DataFrame(dict(actual=labels_test, prediction = preds))
pd.crosstab(index=combined["actual"], columns=combined["prediction"])
```

Out of a 135 matches that were a loss or draw we predicted 81 correctly and 44 incorrectly
Out of the 69 matches that were a win we predicted 30 correctly and 39 incorrectly

```python
from sklearn.metrics import precision_score

precision_score(labels_test, preds)
```

```python
def rolling_avg(group, cols, new_cols):
    group = group.sort_values("Date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
epl_df.columns
```

```python
cols = ["GF", "GA", "Sh", "SoT", "Att 3rd", "Mid 3rd", "Dist", "PK"]
new_cols = [f"{x}_rolling" for x in cols]
new_cols
```

```python
epl_df_rolling = epl_df.groupby("Club").apply(lambda x: rolling_avg(x, cols, new_cols))
```

```python
epl_df_rolling
```

```python
epl_df_rolling
```

```python
epl_df_rolling.index = range(epl_df_rolling.shape[0])
```

```python
epl_df_rolling.shape
```

```python
def make_predictions(data, features):
    train = data[data["Date"] < '2024-01-01']
    test = data[data["Date"] >= '2024-01-01']
    rf.fit(train[features], train["Target"])
    preds = rf.predict(test[features])
    combined = pd.DataFrame(dict(actual=test["Target"], predicted=preds), index=test.index)
    accuracy = accuracy_score(preds, test["Target"])
    precision = precision_score(preds, test["Target"])
    return combined, precision, accuracy
      
```

```python
predictors = predictors + new_cols
combined, precision, accuracy = make_predictions(epl_df_rolling, predictors)
print(accuracy)
```
